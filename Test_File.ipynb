```python
import pandas as pd
import json
import os

path = r"C:\Users\Hari\Desktop\pip\pulse\data\aggregated\transaction\country\india\state"
Agg_state_list = os.listdir(path)

clm = {'State': [],'Year': [],'Quarter': [],'Transaction_type': [],'Transaction_count': [],'Transaction_amount': []}

for state in Agg_state_list:                                  
    state_path = os.path.join(path, state)
    year_list = os.listdir(state_path)

    for year in year_list:                                   
        year_path = os.path.join(state_path, year)
        quarter_list = os.listdir(year_path)

        for q in quarter_list:                                
            if q.endswith(".json"):
                file_path = os.path.join(year_path, q)
                with open(file_path, "r") as f:
                    data = json.load(f)
          
                    for item in data["data"]["transactionData"]:
                        name = item["name"]
                        count = item["paymentInstruments"][0]["count"]
                        amount = item["paymentInstruments"][0]["amount"]

                        clm['State'].append(state)
                        clm['Year'].append(year)
                        clm['Quarter'].append(int(q.strip(".json")))
                        clm['Transaction_type'].append(name)
                        clm['Transaction_count'].append(count)
                        clm['Transaction_amount'].append(amount)


Agg_Trans= pd.DataFrame(clm)
Agg_Trans
```

```python
import pandas as pd
import json
import os

path = r"C:\Users\Hari\Desktop\pip\pulse\data\map\transaction\hover\country\india\state\\"
map_state_list = os.listdir(path)

clm = {'State': [],'Year': [],'Quarter': [],'District': [],'transaction_type': [],'transaction_count': [],'transaction_amount': []}

for i in map_state_list:                 
    p_i = path + i + "/"
    years = os.listdir(p_i)

    for j in years:                     
        p_j = p_i + j + "/"
        quarters = os.listdir(p_j)

        for k in quarters:              
            if k.endswith(".json"):
                p_k = p_j + k
                with open(p_k, "r") as f:
                    D = json.load(f)

                hover_list = D["data"]["hoverDataList"]

                for z in hover_list:
                    district = z["name"]
                    metric = z["metric"][0]

                    type = metric["type"]
                    count = metric["count"]
                    amount = metric["amount"]

                    clm['State'].append(i)
                    clm['Year'].append(j)
                    clm['Quarter'].append(int(k.replace(".json", "")))
                    clm['District'].append(district)
                    clm['transaction_type'].append(type)
                    clm['transaction_count'].append(count)
                    clm['transaction_amount'].append(amount)

map_trans = pd.DataFrame(clm)
map_trans
```

```python
import pandas as pd
import json
import os

path = r"C:\Users\Hari\Desktop\pip\pulse\data\top\transaction\country\india\state\\"
top_state_list = os.listdir(path)

clm2 = {'State': [],'Year': [],'Quarter': [],'transaction_type': [],'transaction_count': [],'transaction_amount': [],'transaction_pincode': []}

for i in top_state_list:                            
    p_i = path + i + "/"
    if not os.path.isdir(p_i):                      
        continue
    year_list = os.listdir(p_i)

    for j in year_list:                             
        p_j = p_i + j + "/"
        if not os.path.isdir(p_j):
            continue
        quarter_list = os.listdir(p_j)

        for k in quarter_list:                      
            if not k.endswith(".json"):
                continue
            p_k = p_j + k
            with open(p_k, "r") as Data:
                D = json.load(Data)

            for z in (D.get("data", {}).get("districts") or []):
                clm2['transaction_type'].append(z["entityName"])
                clm2['transaction_count'].append(z["metric"]["count"])
                clm2['transaction_amount'].append(z["metric"]["amount"])
                clm2['transaction_pincode'].append(None)

                clm2['State'].append(i)
                clm2['Year'].append(j)
                clm2['Quarter'].append(int(k.replace(".json", "")))

            for z in (D.get("data", {}).get("pincodes") or []):
                clm2['transaction_type'].append(None)
                clm2['transaction_count'].append(z["metric"]["count"])
                clm2['transaction_amount'].append(z["metric"]["amount"])
                clm2['transaction_pincode'].append(z["entityName"])

                clm2['State'].append(i)
                clm2['Year'].append(j)
                clm2['Quarter'].append(int(k.replace(".json", "")))

Top_trans = pd.DataFrame(clm2)
Top_trans
```

```python
import pandas as pd
import json
import os

path = r"C:\Users\Hari\Desktop\pip\pulse\data\top\user\country\india\state"
state_list = os.listdir(path)

clm = {'State': [],'Year': [],'Quarter': [],'user_district': [],'user_pincode': [],'registered_users': []}

for state in state_list:
    p_state = os.path.join(path, state)
    if not os.path.isdir(p_state):
        continue
    year_list = os.listdir(p_state)

    for year in year_list:
        p_year = os.path.join(p_state, year)
        if not os.path.isdir(p_year):
            continue
        quarter_list = os.listdir(p_year)

        for q in quarter_list:
            if not q.endswith(".json"):
                continue
            p_q = os.path.join(p_year, q)
            with open(p_q, "r") as f:
                D = json.load(f)
            quarter = int(q.replace(".json", ""))

            for d in D["data"].get("districts", []):
                clm['State'].append(state)
                clm['Year'].append(year)
                clm['Quarter'].append(quarter)

                clm['user_district'].append(d["name"])
                clm['user_pincode'].append(None)
                clm['registered_users'].append(d["registeredUsers"])

            for p in D["data"].get("pincodes", []):
                clm['State'].append(state)
                clm['Year'].append(year)
                clm['Quarter'].append(quarter)

                clm['user_district'].append(None)
                clm['user_pincode'].append(p["name"])
                clm['registered_users'].append(p["registeredUsers"])


Top_user = pd.DataFrame(clm)
Top_user
```
